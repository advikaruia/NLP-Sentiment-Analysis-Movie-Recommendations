```python
import numpy as np
import pandas as pd
```

```python
# Read in our positive data
filepath = "../Data/positive-training-reviews.csv"

df_pos = pd.read_csv(filepath)
df_pos.head()
```

```python
# Read in our negative data
filepath = "../Data/negative-training-reviews.csv"

df_neg = pd.read_csv(filepath)
df_neg.head()
```

```python
# We will now merge the two datasets together
df = pd.DataFrame()
df = df_pos
df.head()
```

```python
df = df.merge(df_neg, how='inner', on='vocab')
```

```python
df.head()
```

```python
target = np.zeros(2002) # target vector
```

```python
for n, i in enumerate(df.columns[1:]):
    if i.startswith("p"):
        target[n] = 1
    elif i.startswith("n"):
        target[n] = 0
    else:
        continue
```

```python
target
```

```python
df = df.set_index("vocab")
df.head()
```

```python
df.to_numpy().T
```

```python
X = df.to_numpy().T
len(X)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

# train and test data
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.3)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, y_train)
```

```python
LR.score(X_train, y_train), LR.score(X_test, y_test)
```

**Therefore, after testing our trained model with the test data, we arrived at an accuracy score of 84.526%.**
